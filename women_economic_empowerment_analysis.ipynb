```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Age": [24, 31, 28, 35, 22, 40, 29, 33, 26, 38],
    "County": ["Nakuru", "Kisumu", "Nairobi", "Mombasa", "Nakuru", "Eldoret", "Kisumu", "Nairobi", "Mombasa", "Nakuru"],
    "Income": [18000, 25000, 30000, 22000, 15000, 35000, 27000, 32000, 20000, 40000],
    "Savings": [3000, 5500, 7000, 4000, 2000, 9000, 6000, 8000, 3500, 10000],
    "Business_Type": ["Retail", "Farming", "Salon", "Retail", "Tailoring", "Farming", "Retail", "Salon", "Tailoring", "Retail"],
    "Smartphone": ["Yes", "No", "Yes", "Yes", "No", "Yes", "Yes", "Yes", "No", "Yes"],
    "Monthly_Health_Cost": [1500, 2300, 1800, 2100, 1200, 2500, 1900, 2000, 1400, 2600]
}

df = pd.DataFrame(data)
```

```python
average_income = df["Income"].mean()

print("Average Income:", average_income)
```

```python
average_savings = df["Savings"].mean()

print("Average Savings:", average_savings)
```

```python
highest_income = df["Income"].max()
lowest_income = df["Income"].min()

print("Highest Income:", highest_income)
print("Lowest Income:", lowest_income)
```

```python
df["County"].value_counts()
```

```python
df["Smartphone"].value_counts()
```

```python
df.groupby("Business_Type")["Income"].mean()
```

```python
df.groupby("County")["Savings"].mean()
```

```python
average_health_cost = df["Monthly_Health_Cost"].mean()

print("Average Monthly Health Cost:", average_health_cost)
```

```python
df.groupby("County")["Income"].mean().plot(kind="bar")

plt.title("Average Income by County")
plt.xlabel("County")
plt.ylabel("Average Income (KES)")
plt.show()
```

```python
df["Business_Type"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Distribution of Business Types")
plt.ylabel("")
plt.show()
```

```python
df["Monthly_Health_Cost"].plot(kind="hist", bins=5)

plt.title("Distribution of Monthly Health Costs")
plt.xlabel("Monthly Health Cost (KES)")
plt.ylabel("Number of Participants")
plt.show()
```

```python
df["Smartphone"].value_counts().plot(kind="bar")

plt.title("Smartphone Ownership")
plt.xlabel("Smartphone")
plt.ylabel("Number of Participants")
plt.show()
```

```python
print("RESEARCH INSIGHTS")
print("-----------------")
print("Average Income:", df["Income"].mean())
print("Average Savings:", df["Savings"].mean())
print("Average Monthly Health Cost:", df["Monthly_Health_Cost"].mean())
print("Highest Income:", df["Income"].max())
print("Lowest Income:", df["Income"].min())
```

```python
above_average = df[df["Income"] > df["Income"].mean()]

above_average
```

## Key Findings

Based on the analysis of the sample survey dataset:

- The average monthly income of participants was approximately **KES 26,300**.
- The average monthly savings were approximately **KES 5,800**.
- Participants from **Eldoret** recorded the highest average income, followed by **Nairobi**.
- **Retail** was the most common business activity among the participants.
- Most participants reported owning a smartphone, indicating relatively high digital access.
- Monthly healthcare expenses varied across participants, highlighting the financial impact of healthcare costs.
