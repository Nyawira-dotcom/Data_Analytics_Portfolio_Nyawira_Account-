```python
# ===========================================
# M-Pesa Transaction Processor
# This program processes M-Pesa transactions,
# categorises them and generates a summary report.
# ===========================================

# List of M-Pesa transaction amounts in KES
# Added three extra amounts:
# -300 (invalid), 160000 (above limit), 50000 (large transaction)

transactions = [
    150, 4500, 12000, 800, 25000,
    3200, 600, 18000, 950, 7500,
    -300, 160000, 50000
]

# Counters
small_count = 0
medium_count = 0
large_count = 0

# Variables for calculations
total = 0
highest = None
lowest = None

print("Processing transactions...")
print("-----------------------------------------")

# Process each transaction
for amount in transactions:

    # Check for invalid transaction
    if amount < 0:
        print(f"Invalid transaction: KES {amount:,}")
        continue

    # Check if transaction exceeds M-Pesa daily limit
    if amount > 150000:
        print(f"Flagged transaction: KES {amount:,}")
        continue

    # Add to total
    total += amount

    # Find highest transaction
    if highest is None or amount > highest:
        highest = amount

    # Find lowest transaction
    if lowest is None or amount < lowest:
        lowest = amount

    # Categorise transaction
    if amount < 1000:
        category = "Small"
        small_count += 1
    elif amount < 10000:
        category = "Medium"
        medium_count += 1
    else:
        category = "Large"
        large_count += 1

    # Display each transaction and its category
    print(f"Transaction: KES {amount:,} → Category: {category}")

# Calculate totals
valid_transactions = small_count + medium_count + large_count
average = total / valid_transactions

# Display summary report
print("\n====== M-PESA TRANSACTION SUMMARY ======")
print(f"Total Transactions : {valid_transactions}")
print(f"Total Amount : KES {total:,}")
print(f"Average Amount : KES {average:,.2f}")
print(f"Highest Transaction : KES {highest:,}")
print(f"Lowest Transaction : KES {lowest:,}")
print("-----------------------------------------")
print(f"Small (<1,000) : {small_count} transactions")
print(f"Medium (1K - 10K) : {medium_count} transactions")
print(f"Large (>=10,000) : {large_count} transactions")
print("=========================================")

# While loop to allow the user to check more amounts
while True:
    answer = input("\nCheck another amount? (yes/no): ").lower()

    if answer == "no":
        print("Thank you for using M-Pesa Processor. Goodbye!")
        break

    elif answer == "yes":
        amount = float(input("Enter amount (KES): "))

        if amount < 0:
            print(f"Invalid transaction: KES {amount:,.0f}")

        elif amount > 150000:
            print(f"Flagged transaction: KES {amount:,.0f}")

        elif amount < 1000:
            print(f"Transaction: KES {amount:,.0f} → Category: Small")

        elif amount < 10000:
            print(f"Transaction: KES {amount:,.0f} → Category: Medium")

        else:
            print(f"Transaction: KES {amount:,.0f} → Category: Large")

    else:
        print("Please enter either 'yes' or 'no'.")
```
