```python
# ==========================================================
#              DATA PROCESSING TOOLKIT
# ==========================================================

print("=" * 60)
print("              DATA PROCESSING TOOLKIT")
print("=" * 60)

# Required Test Datasets
dataset1 = [5, 10, 15, 20]
dataset2 = [100, 110, 95, 90]
dataset3 = [8, 9, 10, 250]


# -------------------------
# Calculate Mean
# -------------------------
def calculate_mean(data):
    return sum(data) / len(data)


# -------------------------
# Calculate Median
# -------------------------
def calculate_median(data):
    numbers = sorted(data)
    n = len(numbers)

    if n % 2 == 0:
        middle1 = numbers[n//2 - 1]
        middle2 = numbers[n//2]
        return (middle1 + middle2) / 2
    else:
        return numbers[n//2]


# -------------------------
# Count Values
# -------------------------
def count_values(data):
    return len(data)


# -------------------------
# Calculate Range
# -------------------------
def calculate_range(data):
    return max(data) - min(data)


# -------------------------
# Find Outliers
# (Values greater than twice the mean)
# -------------------------
def find_outliers(data):
    mean = calculate_mean(data)
    outliers = []

    for value in data:
        if value > mean * 2:
            outliers.append(value)

    return outliers


# -------------------------
# Generate Report
# -------------------------
def generate_report(data, name):

    print("\n" + "=" * 60)
    print(name)
    print("=" * 60)

    print("Dataset :", data)
    print("Count   :", count_values(data))
    print("Mean    : {:.2f}".format(calculate_mean(data)))
    print("Median  :", calculate_median(data))
    print("Highest :", max(data))
    print("Lowest  :", min(data))
    print("Range   :", calculate_range(data))

    outliers = find_outliers(data)

    if len(outliers) == 0:
        print("Outliers: None")
    else:
        print("Outliers:", outliers)


# -------------------------
# Compare Datasets
# -------------------------
def compare_datasets(data1, data2):

    print("\n" + "=" * 60)
    print("DATASET COMPARISON")
    print("=" * 60)

    mean1 = calculate_mean(data1)
    mean2 = calculate_mean(data2)

    print("Dataset 1 Mean :", round(mean1,2))
    print("Dataset 2 Mean :", round(mean2,2))

    if mean1 > mean2:
        print("Higher Mean : Dataset 1")
    elif mean2 > mean1:
        print("Higher Mean : Dataset 2")
    else:
        print("Both datasets have the same mean.")

    print()

    print("Dataset 1 Size :", count_values(data1))
    print("Dataset 2 Size :", count_values(data2))

    if count_values(data1) > count_values(data2):
        print("Larger Dataset : Dataset 1")
    elif count_values(data2) > count_values(data1):
        print("Larger Dataset : Dataset 2")
    else:
        print("Both datasets are the same size.")


# -------------------------
# Run Reports
# -------------------------

generate_report(dataset1, "DATASET 1")
generate_report(dataset2, "DATASET 2")
generate_report(dataset3, "DATASET 3")

# -------------------------
# Compare Datasets
# -------------------------

compare_datasets(dataset1, dataset2)
compare_datasets(dataset2, dataset3)
compare_datasets(dataset1, dataset3)

print("\n" + "=" * 60)
print("PROJECT COMPLETED SUCCESSFULLY")
print("=" * 60)
```
